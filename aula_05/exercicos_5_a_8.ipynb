#Exercícios Individuais – Filtragem e Classificação

```python
import pandas as pd
```

```python
df = pd.read_json("/content/vinhos.json", lines=True)
df
```

##5 – Crie dois dataFrames, um com o grape_type Wine no location_type Brasil e outro com o grape_type Table todos da category Sales. No fim, use dicribe().T


```python
df1 = df.copy()
df1 = df1[(df1["grape_type"] == "Wine") & (df1["location_type"] == "Brazil_State") & (df1["category"] == "Sales")]  # Selects rows where grape_type is Wine
df1
```

```python
df1.describe().T
```

```python
df2 = df.copy()
df2 = df2[(df2["grape_type"] == "Table") & (df2["category"] == "Sales")]  # Selects rows where grape_type is Wine
df2
```

```python
df2.describe().T
```

##6 – Nestes dataFrames, ordene o ano de forma decrescente e o value_usd crescente. No fim, selecione as colunas year, grape_type, volume_hl, value_usd e location, nessa ordem. Use iloc[]

```python
df1.sort_values(by=["year", "volume_hl"], ascending=[False, True], inplace=True)
df1
```

```python
df1.iloc[:,[0,5,6,7,2]]
```

```python
df2.sort_values(by=["year", "volume_hl"], ascending=[False, True], inplace=True)
df2
```

```python
df2.iloc[:,[0,5,6,7,2]]
```
