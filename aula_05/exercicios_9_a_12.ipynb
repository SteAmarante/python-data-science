##11 – Crie o dfFinal, a partir do df3, realizando a retirada dos duplcicados, monstrando o shape do antes e depois. Por fim subistitua os Vazios de cada coluna (para volume_hl e value_usd, substitua usando a mediana median()).


```python
dfFinal = df3.copy()
dfFinal.shape
```

```python
dfFinal = dfFinal.drop_duplicates()
dfFinal.shape
```

```python
mean_volume = dfFinal['volume_hl'].mean()
dfFinal['volume_hl'] = dfFinal['volume_hl'].fillna(mean_volume)
mean_value = dfFinal['value_usd'].mean()
dfFinal['value_usd'] = dfFinal['value_usd'] .fillna(mean_value)
dfFinal.head(20)
```
