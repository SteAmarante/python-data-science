#AULA 8 – Frequência · Tendência Central

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/vinhos.csv")
df
```

Categoricas e Númericas discretas: devemos obter as frequências para cada valor dessa variável.

```python
freq_cat = df["category"].value_counts()
rel_cat = df["category"].value_counts(normalize=True)*100
acum_cat = freq_cat.cumsum()
tabela = pd. concat([freq_cat, rel_cat, acum_cat], axis=1)
tabela.columns = ["Freq", "Rel_%", "Acum"]
print(tabela)
```

Númericas contínuas: devemos primeiro construir intervalos e depois obter as frequências para cada intervalo.

```python
# 2. Definir FAIXAS (bins) para a variável contínua
bins = pd.interval_range(start=30, end=900, freq=145)

# 3. Classificar cada preço no respectivo intervalo
df["faixa_preco"] = pd.cut(df["value_usd"], bins=bins, right=False)

# 4. Construir tabela de frequência
freq_abs = df["faixa_preco"].value_counts().sort_index()
freq_rel = (freq_abs / len(df)) * 100
freq_acum = freq_abs. cumsum()

tabela = pd.DataFrame({
    "Faixa de preço": freq_abs.index,
    "Frequência": freq_abs.values,
    "Relativa (%)": freq_rel.round(2).values,
    "Acumulada": freq_acum.values})


tabela
```
