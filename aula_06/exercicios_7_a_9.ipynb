#Média, Moda e Mediana


```python
import pandas as pd
```

```python
df = pd.read_csv("/content/vinhos.csv")
df
```

##1 - Insira um outlier value_usd=500000 em numa cópia do DF de vinhos; observe mudança em média e mediana


```python
df2 = df.copy() #faz uma cópia da tabela original
df2.loc[2:2, 'value_usd'] = 500000 #você também pode usar o iloc colocando a linha e a coluna que vc quer trocar o valor.
df2.head()
```

```python
#Original
media_og = df['value_usd'].mean()
media_og
```

```python
#Cópia
media_value = df2['value_usd'].mean()
media_value
```

```python
#Original
media_og = df['value_usd'].median()
media_og
```

```python
#Cópia
media_value = df2['value_usd'].median()
media_value
```

##2 - Calcule moda de **year** usando toda a base, calcule a **moda** novamente usando somente uma amostra de 10%

```python
#Moda total
df["year"].mode()
```

```python
#moda da amostra
amostra = df.sample(frac=0.1, random_state=42)
amostra["year"].mode()
```

##3 - Calcule média e mediana de **value_usd por grape_type** (use groupby).

```python
#Df original
media_geral = df.groupby('grape_type')['value_usd'].mean()
mediana_geral = df.groupby('grape_type')['value_usd'].median()

tabela_geral = pd.DataFrame({'Média': media_geral, 'Mediana': mediana_geral})
tabela_geral
```

```python
#Df Copia
media_geral_2 = df2.groupby('grape_type')['value_usd'].mean()
mediana_geral_2 = df2.groupby('grape_type')['value_usd'].median()

tabela_geral_2 = pd.DataFrame({'Média': media_geral_2, 'Mediana': mediana_geral_2})
tabela_geral_2
```
