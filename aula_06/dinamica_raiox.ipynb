##3. Montar um dataFrame com a classificaçao do tipo de variável (Categórica nominal/ordinal ou numerica continua/discreta) de cada coluna.


CÓDIGO	DESCRIÇÃO	UNIDADE	UF	Preço.Entregue	Preço.Final	REF	PEI	Relação.Liderança	Item.Líder	Família	TipoInsumo	Nota.Revisional	Região

```python
types_df = pd.DataFrame({
    'coluna nome': amostra_1.columns,
    'Tipo variavel': amostra_1.dtypes
})
types_df.loc[types_df['Tipo variavel'] == 'object', 'Tipo variavel'] = 'categorica'
types_df.loc[types_df['Tipo variavel'] == 'float', 'Tipo variavel'] = 'numerica'

display(types_df)
```

```python
types_df1 = pd.DataFrame({
    'coluna nome': amostra_2.columns,
    'Tipo variavel': amostra_2.dtypes
})
types_df1.loc[types_df1['Tipo variavel'] == 'object', 'Tipo variavel'] = 'categorica nominal'
types_df1.loc[types_df1['Tipo variavel'] == 'float', 'Tipo variavel'] = 'numerica continuo'
types_df1.loc[types_df1['coluna nome'] == 'Relação.Liderança', 'Tipo variavel'] = 'categorica ordinal'
types_df1.loc[types_df1['Tipo variavel'] == 'int', 'Tipo variavel'] = 'numerica discreta'
display(types_df1)
```

##4. Nas amostras calcular a frequência absoluta e relativa (%) da coluna Tipolnsumo.

```python
# Frequência absoluta
freq_abs = df["TipoInsumo"].value_counts()

# Frequência relativa em %
freq_rel = df["TipoInsumo"].value_counts(normalize=True) * 100

# Mostrar juntas em um DataFrame
tabela_freq = pd.DataFrame({
    "Frequência Absoluta": freq_abs,
    "Frequência Relativa (%)": freq_rel
})

print(tabela_freq)
```

##5. Calcular as medidas de tendencia central da tabela da regiao de foco (Média, Mediana e Moda) total e AGRUPANDO por UF (Só Média e Mediana) e por Tipo Insumo (Só Média e Mediana).


```python
modauf = amostra_1.groupby('UF')['Relação.Liderança'].agg(
    mode=lambda x:pd.Series.mode(x)
)
display(modauf)
```

```python
modauf2 = amostra_2.groupby('UF')['Relação.Liderança'].agg(
    mode=lambda x:pd.Series.mode(x)
)
display(modauf2)
```

```python
modatipo = amostra_1.groupby('TipoInsumo')['Família'].agg(
    mode=lambda x:pd.Series.mode(x)
)
display(modatipo)
```

```python
modatipo2 = amostra_2.groupby('TipoInsumo')['Família'].agg(
    mode=lambda x:pd.Series.mode(x)
)
display(modatipo2)
```

AGRUPANDO por UF (Só Média e Mediana)

```python
uf = amostra_1.groupby('UF')['Preço.Final'].agg(
    mean='mean',
    median='median',
    )
display(uf)
```

```python
uf2 = amostra_2.groupby('UF')['Preço.Final'].agg(
    mean='mean',
    median='median',
    )
display(uf2)
```

Tipo Insumo (Só Média e Mediana).

```python
tipo_insumo = amostra_1.groupby('TipoInsumo')['Preço.Final'].agg(
    mean='mean',
    median='median',
)
display(tipo_insumo)
```

```python
tipo_insumo2 = amostra_2.groupby('TipoInsumo')['Preço.Final'].agg(
    mean='mean',
    median='median',
)
display(tipo_insumo2)
```

## Insights da Análise

Com base na análise exploratória dos dados das duas amostras de 25% da região Sudeste do SICRO, podemos destacar os seguintes insights:

* **Qualidade dos Dados:** Ambas as amostras apresentaram um número pequeno de valores nulos, principalmente nas colunas 'Item.Líder' e 'Região'. A exclusão dessas linhas nulas resultou em amostras limpas para análise.
* **Tipos de Variáveis:** A maioria das colunas no dataset é categórica nominal, com exceção das colunas 'Preço.Entregue' e 'Preço.Final' que são numéricas contínuas, e 'Relação.Liderança' que foi classificada como categórica ordinal.
* **Distribuição dos Tipos de Insumo:** A análise da frequência dos tipos de insumo na amostra 1 (que reflete a distribuição do dataset completo) mostrou que 'Material DMT' é o tipo de insumo mais frequente, seguido por 'Equipamento' e 'Liderado'.
* **Medidas de Tendência Central por UF:** Ao agrupar por UF, observamos variações na média e mediana dos preços finais entre os estados. Isso sugere que os custos de insumos podem variar significativamente dentro da própria região Sudeste.
* **Medidas de Tendência Central por Tipo de Insumo:** A análise por tipo de insumo revelou grandes diferenças entre as médias e medianas. Equipamentos, por exemplo, apresentam médias e medianas muito mais altas do que outros tipos de insumo, como Material DMT ou Liderado. A moda para a maioria dos tipos de insumo e UFs foi 'Sem Família' e 'LÍDER', respectivamente, indicando a prevalência de itens sem classificação familiar específica e itens que são líderes em suas relações de liderança.

Esses insights iniciais fornecem uma base para análises mais aprofundadas, como a comparação de preços entre UFs para tipos de insumo específicos ou a investigação dos insumos com preços mais elevados.
