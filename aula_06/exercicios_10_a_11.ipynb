#Medidas de Dispersão e Padronização

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/vinhos.csv")
df
```

##10. Calcule a amplitude absoluta, relativa, variança e desvio padrão de volume_hl por grape_type. Qual a interpretaçao que teremos?


```python
#amplitude absoluta e relativa
amplitude_absoluta = df.groupby("grape_type")["volume_hl"].max() - df.groupby("grape_type")["volume_hl"].min()
amplitude_relativa = amplitude_absoluta / df.groupby("grape_type")["volume_hl"].max()

#variança
varianca = df.groupby("grape_type")["volume_hl"].var()

#desvio padrão
desvio_padrao = df.groupby("grape_type")["volume_hl"].std()

tabela_calculo = pd.DataFrame({
    "Amplitude Absoluta": amplitude_absoluta,
    "Amplitude Relativa": amplitude_relativa,
    "Variância": varianca,
    "Desvio Padrão": desvio_padrao
})

tabela_calculo
```

##11. Selecione uma amostra de 25%, e calcule as medidas de dispersão para value_usd e volume_hl por grape_type. Compare-as com as da populacao. Por que diferem?

```python
amostra = df.sample(frac=0.25, random_state=42)
amostra_amplitude_absoluta = amostra.groupby("grape_type")["value_usd"].max() - amostra.groupby("grape_type")["volume_hl"].min()
amostra_amplitude_relativa = amostra_amplitude_absoluta / amostra.groupby("grape_type")["volume_hl"].max()

#variança
amostra_varianca = amostra.groupby("grape_type")["value_usd"].var()
amostra_varianca_2 = df.groupby("grape_type")["volume_hl"].var()

#desvio padrão
amostra_desvio_padrao = amostra.groupby("grape_type")["value_usd"].std()

tabela_calculo_2 = pd.DataFrame({
    "Amplitude Absoluta": amostra_amplitude_absoluta,
    "Amplitude Relativa": amostra_amplitude_relativa,
    "Variância Amostral": amostra_varianca,
    "Variância Amostral 2": amostra_varianca_2,
    "Desvio Padrão": desvio_padrao
})

tabela_calculo_2
```

```python
amostra_2 = df.sample(frac=0.25, random_state=42)

print("\n ------- Amostra de 25% de Value_usd ---------- \n")

#amplitude

amp_price = amostra_2["value_usd"].max() - amostra_2["value_usd"].min()
amp_rel = amp_price / amostra_2["value_usd"].mean() * 100

print(f'Amplitude de value_usd: {round(amp_price, 2)}, \nAmplitude relativa de value_usd: {round(amp_rel, 2)}')

#variança

var_price = amostra_2["value_usd"].var()

print(f'\nVariância de value_usd: {round(var_price, 2)}')

#Desvio Padrao
desv_price = amostra_2["value_usd"].std()

print(f'\nDesvio Padrão de value_usd: {round(desv_price, 2)}')
```

```python
print("\n ------- Amostra de 25% de Volume_hl ---------- \n")

#amplitude

amp_price = amostra_2["volume_hl"].max() - amostra_2["volume_hl"].min()
amp_rel = amp_price / amostra_2["volume_hl"].mean() * 100

print(f'Amplitude de volume_hl: {round(amp_price, 2)}, \nAmplitude relativa de volume_hl: {round(amp_rel, 2)}')

#variança

var_price = amostra_2["volume_hl"].var()

print(f'\nVariância de volume_hl: {round(var_price, 2)}')

#Desvio Padrao
desv_price = amostra_2["volume_hl"].std()

print(f'\nDesvio Padrão de volume_hl: {round(desv_price, 2)}')
```
