```python
df = pd.read_csv("/content/vinhos.csv")
df
```

##1 - Sorteie 5 amostras de 5 % cada uma; registre médias de volume_hl.


```python
media_volume = df["volume_hl"].mean()

for i in range(5):
  amostra = df.sample(frac=0.05, random_state=i)
  media_amostra = amostra["volume_hl"].mean()
  print(f" A média do volume é: {round(media_volume,2)}, e a média da amostra é {round(media_amostra, 2)}\n")
```
