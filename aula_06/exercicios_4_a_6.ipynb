#Exercicios - distribuição de Frequência

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/vinhos.csv")
df
```

##4 - Construa uma tabela contendo frequência absoluta, relativa (%) e acumulada da coluna grape_type.



```python
#primeiro modelo
freq_grape = df["grape_type"].value_counts()
rel_grape = df["grape_type"].value_counts(normalize=True)
acum_grape = freq_grape.cumsum()
tabela = pd.concat([freq_grape, rel_grape, acum_grape], axis=1)
tabela.columns = ["Freq", "Rel_%", "Acum"]
print(tabela)
```

```python
#segundo modelo
freq_grape = df["grape_type"].value_counts()
rel_grape = df["grape_type"].value_counts(normalize=True)
acum_grape = freq_grape.cumsum()

tabela_uva = pd.DataFrame({
    "Freq": freq_grape,
    "Rel_%": rel_grape.round(2),
    "Acum": acum_grape
})

print("\n1) Tabela de grape_type:")
tabela_uva
```

##5 - Use pd.qcut para dividir value_usd em 4 classes iguais (quartis) e gere uma tabela com frequência absoluta e relativa de cada quartil. Use o parâmetro q=4.



```python
# 3. Classificar cada preço no respectivo intervalo
df["faixa_preco"] = pd.qcut(df["value_usd"], q=4)

# 4. Construir tabela de frequência
freq_abs = df["faixa_preco"].value_counts().sort_index()
freq_rel = (freq_abs / len(df)) * 100
freq_acum = freq_abs. cumsum()

tabela = pd.DataFrame({
    "Faixa de preço": freq_abs.index,
    "Frequência": freq_abs.values,
    "Relativa (%)": freq_rel.round(2).values,
    "Acumulada": freq_acum.values})

tabela
```

```python
#segundo modelo

quartis = pd.qcut(df["value_usd"], q=4)
freq_q_abs = quartis.value_counts()
freq_q_rel = (freq_q_abs / len(df) * 100).round(2)

tabela_quartis = pd.DataFrame({
    "Faixa de preço": freq_q_abs.index,
    "Frequência": freq_q_abs.values,
    "Relativa (%)": freq_q_rel.values
})

print("\n2) Frequencia por quartis de price:")
tabela
```

##6 - Agrupe os anos em três intervalos (2015–2017, 2018–2020, 2021–2024). Para cada faixa, calcule a soma de volume_hl e apresente a frequência relativa (%) desses totais em relação ao volume global do dataset.volume_tot_faixa / volume_global

```python
bins_ano = [2015, 2018, 2021, 2025] #2025 é o limite superior exclusivo
df["faixa_ano"] = pd.cut(df["year"], bins=bins_ano, right=False)

volume_tot_faixa = df.groupby("faixa_ano")["volume_hl"].sum()
volume_global = df["volume_hl"].sum()
freq_volume_rel = (volume_tot_faixa / volume_global * 100).round(2)

tabela_volume = pd.DataFrame({
    "Volume total": volume_tot_faixa,
    "Relativa (%)": freq_volume_rel
})

print("\n4) Volume de wine por faixa anual (freq. relativa)")
tabela_volume
```
