#Raio‑X das tabelas de preço públicas do SICRO

**Cenário**

Descobrir quais tipos de insumo apresentam maior dispersão de preços na sua região‑foco e sugerir causas para essa variação.

**Atividade**

Nas mesmas duas amostras de 25% com seeds diferentes da sua região foco, da dinamica anterior:

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/SICRO_2023_2024_Sudeste.csv", encoding = "latin1")
```

```python
amostra_1 = df.sample(frac=0.25, random_state=7)
amostra_1.head()
```

```python
amostra_1 = amostra_1.dropna()
```

```python
amostra_2 = df.sample(frac=0.25, random_state=24)
amostra_2.head()
```

```python
amostra_2 = amostra_2.dropna()
```

## 1 - Nas amostras calcular a amplitude, amplitude relativa %, variância, desvio‑Padrão. E comparar com o da população.

```python
# --- Cálculos para AMOSTRA 1 ---
amplitude_absoluta_1 = amostra_1['Preço.Final'].max() - amostra_1['Preço.Final'].min()
amplitude_relativa_1 = amplitude_absoluta_1 / amostra_1['Preço.Final'].max()
variancia_1 = amostra_1['Preço.Final'].var()
desvio_padrao_1 = amostra_1['Preço.Final'].std()

tabela_amostra_1 = pd.DataFrame({
    "Amplitude Absoluta": [amplitude_absoluta_1],
    "Amplitude Relativa (%)": [amplitude_relativa_1 * 100],
    "Variância": [variancia_1],
    "Desvio Padrão": [desvio_padrao_1]
}, index=["Amostra 1"]) ##define que essa linha se chamará "Amostra 1"

tabela_amostra_1
```

```python
# --- Cálculos para AMOSTRA 2 ---
amplitude_absoluta_2 = amostra_2['Preço.Final'].max() - amostra_2['Preço.Final'].min()
amplitude_relativa_2 = amplitude_absoluta_2 / amostra_2['Preço.Final'].max()
variancia_2 = amostra_2['Preço.Final'].var()
desvio_padrao_2 = amostra_2['Preço.Final'].std()

tabela_amostra_2 = pd.DataFrame({
    "Amplitude Absoluta": [amplitude_absoluta_2],
    "Amplitude Relativa (%)": [amplitude_relativa_2 * 100],
    "Variância": [variancia_2],
    "Desvio Padrão": [desvio_padrao_2]
}, index=["Amostra 2"]) ##define que essa linha se chamará "Amostra 2""

tabela_amostra_2
```

```python
# --- Cálculos para POPULAÇÃO TOTAL ---
amplitude_absoluta_pop = df['Preço.Final'].max() - df['Preço.Final'].min()
amplitude_relativa_pop = amplitude_absoluta_pop / df['Preço.Final'].max()
variancia_pop = df['Preço.Final'].var()
desvio_padrao_pop = df['Preço.Final'].std()

tabela_populacao = pd.DataFrame({
    "Amplitude Absoluta": [amplitude_absoluta_pop],
    "Amplitude Relativa (%)": [amplitude_relativa_pop * 100],
    "Variância": [variancia_pop],
    "Desvio Padrão": [desvio_padrao_pop]
}, index=["População"]) ##define que essa linha se chamará "População"

tabela_populacao
```

```python
# --- Juntar todas as tabelas ---
tabela_final = pd.concat([tabela_amostra_1, tabela_amostra_2, tabela_populacao])

# Mostrar resultados
tabela_final
```

## 2 - Identificar o IQR e Z-score para achar os outliers. Compare-as com a da população, qual foi a interpretação?

```python
# --- Cálculos para AMOSTRA 1 ---
print("\n=== AMOSTRA 1 ===")
# Quartis e IQR
q1_am1 = amostra_1['Preço.Final'].quantile(0.25)
q3_am1 = amostra_1['Preço.Final'].quantile(0.75)
iqr_am1 = q3_am1 - q1_am1
print(f"IQR: {iqr_am1}\n")

# Z-Score
amostra_1['zscore'] = (amostra_1['Preço.Final'] - amostra_1['Preço.Final'].mean()) / amostra_1['Preço.Final'].std(ddof=0)
print(amostra_1[['Preço.Final', 'zscore']].head())

# Outliers via IQR
lim_inf_am1 = q1_am1 - 1.5 * iqr_am1
lim_sup_am1 = q3_am1 + 1.5 * iqr_am1
outliers_iqr_am1 = amostra_1[(amostra_1['Preço.Final'] < lim_inf_am1) | (amostra_1['Preço.Final'] > lim_sup_am1)]
print(f"\nTotal de outliers (IQR): {len(outliers_iqr_am1)}")

# Outliers via Z-Score
outliers_z_am1 = amostra_1[np.abs(amostra_1['zscore']) > 3]
print(f"\nTotal de outliers (Z-Score): {len(outliers_z_am1)}")
```

```python
# --- Cálculos para AMOSTRA 2 ---
print("\n=== AMOSTRA 2 ===")
# Quartis e IQR
q1_am2 = amostra_2['Preço.Final'].quantile(0.25)
q3_am2 = amostra_2['Preço.Final'].quantile(0.75)
iqr_am2 = q3_am2 - q1_am2
print(f"IQR: {iqr_am2}\n")

# Z-Score
amostra_2['zscore'] = (amostra_2['Preço.Final'] - amostra_2['Preço.Final'].mean()) / amostra_2['Preço.Final'].std(ddof=0)
print(amostra_2[['Preço.Final', 'zscore']].head())

# Outliers via IQR
lim_inf_am2 = q1_am2 - 1.5 * iqr_am2
lim_sup_am2 = q3_am2 + 1.5 * iqr_am2
outliers_iqr_am2 = amostra_2[(amostra_2['Preço.Final'] < lim_inf_am2) | (amostra_2['Preço.Final'] > lim_sup_am2)]
print(f"\nTotal de outliers (IQR): {len(outliers_iqr_am2)}")

# Outliers via Z-Score
outliers_z_am2 = amostra_2[np.abs(amostra_2['zscore']) > 3]
print(f"\nTotal de outliers (Z-Score): {len(outliers_z_am2)}")
```

```python
# --- Cálculos para POPULAÇÃO (todos os dados) ---
print("\n=== POPULAÇÃO ===")
# Quartis e IQR
q1_pop = df['Preço.Final'].quantile(0.25)
q3_pop = df['Preço.Final'].quantile(0.75)
iqr_pop = q3_pop - q1_pop
print(f"IQR: {iqr_pop}\n")

# Z-Score
df['zscore'] = (df['Preço.Final'] - df['Preço.Final'].mean()) / df['Preço.Final'].std(ddof=0)
print(df[['Preço.Final', 'zscore']].head())

# Outliers via IQR
lim_inf_pop = q1_pop - 1.5 * iqr_pop
lim_sup_pop = q3_pop + 1.5 * iqr_pop
outliers_iqr_pop = df[(df['Preço.Final'] < lim_inf_pop) | (df['Preço.Final'] > lim_sup_pop)]
print(f"\nTotal de outliers (IQR): {len(outliers_iqr_pop)}")

# Outliers via Z-Score
outliers_z_pop = df[np.abs(df['zscore']) > 3]
print(f"\nTotal de outliers (Z-Score): {len(outliers_z_pop)}")
```

## 3 - Nas suas amostras calcule identifique os outliers (por Z-score ou IQR) além de usar as medidas de variancia (amplitude e variancia) nas colunas de preço para toda o conjunto depois agrupado por UF e por Tipo Insumo.

```python
#Cálculo Amostra 1
# Preço.Entregue
amp_entregue_1 = amostra_1['Preço.Entregue'].max() - amostra_1['Preço.Entregue'].min()
var_entregue_1 = amostra_1['Preço.Entregue'].var()

# Preço.Final
amp_final_1 = amostra_1['Preço.Final'].max() - amostra_1['Preço.Final'].min()
var_final_1 = amostra_1['Preço.Final'].var()

# Outliers Preço.Entregue
Q1_entregue_1 = amostra_1['Preço.Entregue'].quantile(0.25)
Q3_entregue_1 = amostra_1['Preço.Entregue'].quantile(0.75)
IQR_entregue_1 = Q3_entregue_1 - Q1_entregue_1
outliers_entregue_1 = amostra_1[
    (amostra_1['Preço.Entregue'] < (Q1_entregue_1 - 1.5*IQR_entregue_1)) |
    (amostra_1['Preço.Entregue'] > (Q3_entregue_1 + 1.5*IQR_entregue_1))
]

# Outliers Preço.Final
Q1_final_1 = amostra_1['Preço.Final'].quantile(0.25)
Q3_final_1 = amostra_1['Preço.Final'].quantile(0.75)
IQR_final_1 = Q3_final_1 - Q1_final_1
outliers_final_1 = amostra_1[
    (amostra_1['Preço.Final'] < (Q1_final_1 - 1.5*IQR_final_1)) |
    (amostra_1['Preço.Final'] > (Q3_final_1 + 1.5*IQR_final_1))
]

# Cria tabela de resultados
resultados_amostra1 = pd.DataFrame({
    'Amostra': ['Amostra 1', 'Amostra 1'],
    'Coluna': ['Preço.Entregue', 'Preço.Final'],
    'Amplitude': [amp_entregue_1, amp_final_1],
    'Variância': [var_entregue_1, var_final_1],
    'Qtd Outliers': [len(outliers_entregue_1), len(outliers_final_1)]
})

display(resultados_amostra1)
```

```python
#Análise por UF (usando Amostra 1)

# Lista de UFs
ufs = amostra_1['UF'].unique()

# DataFrame para armazenar resultados
resultados_uf = pd.DataFrame()

for uf in ufs:
    dados_uf = amostra_1[amostra_1['UF'] == uf]

    # Cálculos para Preço.Entregue
    amp_entregue = dados_uf['Preço.Entregue'].max() - dados_uf['Preço.Entregue'].min()
    var_entregue = dados_uf['Preço.Entregue'].var()
    Q1_entregue = dados_uf['Preço.Entregue'].quantile(0.25)
    Q3_entregue = dados_uf['Preço.Entregue'].quantile(0.75)
    IQR_entregue = Q3_entregue - Q1_entregue
    outliers_entregue = dados_uf[
        (dados_uf['Preço.Entregue'] < (Q1_entregue - 1.5*IQR_entregue)) |
        (dados_uf['Preço.Entregue'] > (Q3_entregue + 1.5*IQR_entregue))
    ]

    # Cálculos para Preço.Final
    amp_final = dados_uf['Preço.Final'].max() - dados_uf['Preço.Final'].min()
    var_final = dados_uf['Preço.Final'].var()
    Q1_final = dados_uf['Preço.Final'].quantile(0.25)
    Q3_final = dados_uf['Preço.Final'].quantile(0.75)
    IQR_final = Q3_final - Q1_final
    outliers_final = dados_uf[
        (dados_uf['Preço.Final'] < (Q1_final - 1.5*IQR_final)) |
        (dados_uf['Preço.Final'] > (Q3_final + 1.5*IQR_final))
    ]

    # Adiciona ao DataFrame de resultados
    resultados_uf = pd.concat([resultados_uf, pd.DataFrame({
        'Grupo': [f'UF {uf}', f'UF {uf}'],
        'Coluna': ['Preço.Entregue', 'Preço.Final'],
        'Amplitude': [amp_entregue, amp_final],
        'Variância': [var_entregue, var_final],
        'Qtd Outliers': [len(outliers_entregue), len(outliers_final)]
    })])

display(resultados_uf)
```

## 4 - Escrever os insights que obteve a partir da sua análise.

## Insights da Análise de Variância e Outliers

Com base nos cálculos de amplitude, variância, desvio padrão, IQR e Z-score para as duas amostras e a população total, podemos extrair os seguintes insights:

**Comparação entre Amostras e População:**

*   **Amplitude Absoluta e Relativa:** A amplitude absoluta e relativa para ambas as amostras e a população são as mesmas. Isso indica que, embora as amostras sejam menores, elas capturaram a faixa completa dos preços presentes na população. A amplitude relativa de 100% sugere que o preço mínimo em todas as amostras e na população é 0.
*   **Variância e Desvio Padrão:** As variâncias e desvios padrão das amostras são relativamente próximos aos da população total. Isso demonstra que as amostras são representativas da variabilidade de preços presente nos dados completos. A Amostra 1 apresentou uma variância e desvio padrão ligeiramente maiores, indicando uma dispersão um pouco maior dos preços nesta amostra em comparação com a Amostra 2 e a população.
*   **Outliers (IQR vs. Z-Score):**
    *   O método do IQR identificou um número significativamente maior de outliers em ambas as amostras e na população em comparação com o método do Z-score (com um limiar de 3).
    *   A grande quantidade de outliers identificados pelo IQR sugere que a distribuição dos preços é assimétrica, com uma cauda longa de valores altos. O IQR é mais sensível a essa assimetria do que o Z-score, que assume uma distribuição normal.
    *   O número de outliers identificados por ambos os métodos nas amostras é proporcional ao número de outliers na população, reforçando a representatividade das amostras.

**Análise por UF (usando Amostra 1 e Amostra 2):**

*   Ao analisar a amplitude e a variância dos preços por UF (utilizando ambas as amostras), observamos que há variações na dispersão dos preços entre os estados da região Sudeste (RJ, SP, MG, ES).
*   A quantidade de outliers por UF, conforme identificado pelo método IQR nas amostras, também varia. Isso sugere que a dispersão de preços e a presença de valores extremos podem ser diferentes em cada estado.

**Interpretação Geral:**

*   As amostras selecionadas são boas representações da população total em termos de medidas de dispersão (amplitude, variância, desvio padrão) e identificação de outliers.
*   A alta amplitude relativa e o grande número de outliers identificados pelo IQR indicam que a distribuição dos preços finais no SICRO para a região Sudeste é assimétrica e possui valores extremos significativos.
*   A análise por UF sugere que a variabilidade de preços não é uniforme em toda a região Sudeste, podendo haver fatores específicos de cada estado que influenciam a dispersão dos preços dos insumos.
