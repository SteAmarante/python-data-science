```python

from sqlalchemy import create_engine
import pandas as pd
```

```python
# Conexão (cria o arquivo se não existir)
engine = create_engine("sqlite:///vinhos.db")
```

```python
# Leitura de tabela existente
df = pd.read_sql("SELECT * FROM vinhos", engine)

# Exibir primeiras linhas
df.tail()
```

##Total por país

*   Use Pandas + SQLAlchemy para listar o top 10 quantidade de vinhos por pais de uma amostra, ordenado desc.



```python
amostra = df.sample(frac=0.25, random_state=16)
top_10_paises = amostra.groupby('location').size().sort_values(ascending=False).head(10)
print("Top 10 Locations mais frequentes: ")
print(top_10_paises)
```

##Filtro de preço


*   Selecione todos os vinhos com preço >12000 e salve em DataFrame df_premium
*   Quantos registros há?
* Após isso salve a df_premium no BD. E leia-o para testar


```python
df_premium = df[df['value_usd'] > 12000]
df_premium
```

```python
contagem_premium = len(df_premium)
contagem_premium
```

```python
df_premium.to_sql('df_premium', engine, if_exists='replace', index=False)
```

```python
# Confirmar gravação
pd.read_sql("df_premium", engine).head(10)
```

## Preço médio por categoria


*   Calcule  o preço médio por categoria e grave o DataFrame resultante em uma tabela preco_medio usando to_sql. (Use . reset_index() na hora de calcular a media)




```python
preco_medio = df.groupby('category').agg({'value_usd': 'mean'}).reset_index()
preco_medio.to_sql('preco_medio', engine, if_exists='replace', index=False)
preco_medio
```
