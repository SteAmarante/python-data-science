#Bhaskara com Numpy

Use: np.lib.scimath.sqrt() Para calcular a raiz de delta

```python
import numpy as np
```

```python
#Solicitar ao usuario os coeficientes de a, b e c (numeros reais)

a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))


# Se a == 0, informar ao usuario que não é uma equação do 2º
if a == 0:
  print("Não é uma equação de 2 grau")

# Realizar o calculo de Bhaskara
else:
  delta = (b**2)-(4*a*c)
  raiz_delta = np.lib.scimath.sqrt(delta)
  x1 = (-b + raiz_delta)/(2*a)
  x2 = (-b - raiz_delta)/(2*a)
  print(f"\nO valor de x1 é {x1:.2f} e o valor de x2 {x2:.2f}.")
```
